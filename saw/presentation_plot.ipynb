```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
walk = np.genfromtxt("random_walk_result21_0.txt", dtype=int)
walk_pivot = np.genfromtxt("random_walk_result21_1.txt", dtype=int)
walk_wrong = np.genfromtxt("random_walk_result21_2.txt", dtype=int)
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(walk[:, 0], walk[:,1], color="blue")
upx = max(np.max(walk[:, 0]), np.max(walk_pivot[:, 0]))
downx = min(np.min(walk[:, 0]), np.min(walk_pivot[:, 0]))
upy = max(np.max(walk[:, 1]), np.max(walk_pivot[:, 1]))
downy = min(np.min(walk[:, 1]), np.min(walk_pivot[:, 1]))
mx = [i for i in range(min(downx, downy), max(upx, upy))]
ps = mx
ps.extend([upx, downx, upy, downy])
gridpoints = np.array([(i, j) for i in ps for j in ps])
plt.scatter(gridpoints[:, 0], gridpoints[:, 1],  color="black", s=.1)
plt.xticks(mx)
plt.yticks(mx)
plt.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
plt.annotate("0", (0,0), horizontalalignment="right")
plt.savefig("pivot_move_0.pdf")
plt.scatter(-1, -1,  color="red")
plt.savefig("pivot_move_1.pdf")
plt.plot(walk_pivot[:, 0], walk_pivot[:,1], color="blue", alpha=.5)
plt.savefig("pivot_move_2.pdf")
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(walk_pivot[:, 0], walk_pivot[:,1], color="blue")
upx = max(np.max(walk[:, 0]), np.max(walk_pivot[:, 0]))
downx = min(np.min(walk[:, 0]), np.min(walk_pivot[:, 0]))
upy = max(np.max(walk[:, 1]), np.max(walk_pivot[:, 1]))
downy = min(np.min(walk[:, 1]), np.min(walk_pivot[:, 1]))
mx = [i for i in range(min(downx, downy), max(upx, upy))]
ps = mx
ps.extend([upx, downx, upy, downy])
gridpoints = np.array([(i, j) for i in ps for j in ps])
plt.scatter(gridpoints[:, 0], gridpoints[:, 1],  color="black", s=.1)
plt.xticks(mx)
plt.yticks(mx)
plt.tick_params(labelleft=False, left=False, labelbottom=False, bottom=False)
plt.annotate("0", (0,0), horizontalalignment="right")
plt.savefig("pivot_move_3.pdf")
```
