```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("df3.txt", names = ["N", "square_disp", "modus"])
```

```python
#df = df[df.N>199]
#df = df[df.N>3001]
df = df[df.modus!=1]
```

```python
df = df[df.modus==2]
plt.xscale("log")
plt.yscale("log")
plt.scatter(df["N"], df["square_disp"], c=df["modus"], cmap="Set1", zorder=2, alpha=0.5)
plt.xlabel(r"log($N$)")
plt.ylabel(r"log($\left<x^2\right>$)")
x = np.log(df["N"])
y = np.log(df["square_disp"])
coeff, v = np.polyfit(x[:], y[:], 1, cov=True)
b, a = coeff
b_err = v[0][0]**0.5
plt.plot(np.exp(x), np.exp(a + b*x), zorder=1, label=r"Fit mit $\alpha=$%f +- %f" % (b, b_err))
plt.legend(loc="upper left")
plt.savefig("saw3d.pdf")
```

```python
df = df[df.modus==0]
plt.xscale("log")
plt.yscale("log")
plt.scatter(df["N"], df["square_disp"], c=df["modus"], cmap="Set1", zorder=2, alpha=0.5)
plt.xlabel(r"log($N$)")
plt.ylabel(r"log($\left<x^2\right>$)")
x = np.log(df["N"])
y = np.log(df["square_disp"])
b, a = np.polyfit(x, y, 1)
plt.plot(np.exp(x), np.exp(a + b*x), zorder=1, label=r"Fit mit $\alpha=$%f" % (b))
plt.legend(loc="upper left")
#plt.title("self-avoiding RW in 3d")
plt.savefig("saw3d_abzaehlen.pdf")
```

```python
#plt.xscale("log")
#plt.yscale("log")
plt.scatter(np.log(df["N"]), np.log(df["square_disp"]), c=df["modus"], cmap="Set1", zorder=2, alpha=0.5)
plt.xlabel(r"$\log(N)$")
plt.ylabel(r"$\log(\left<x^2\right>)$")
x = np.log(df["N"])
y = np.log(df["square_disp"])
b, a = np.polyfit(x, y, 1)
plt.plot(x, a + b*x, zorder=1)
plt.legend(["Fit mit Steigung %f" % (b), "grau=MMC, orange=einfaches Monte-Carlo,\n rot=abzhlen"], loc="upper left")
plt.title("self-avoiding RW in 3d")
plt.savefig("saw3d.pdf")
```
