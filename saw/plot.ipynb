```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("df.txt", names = ["N", "square_disp", "mode"])
```

```python
df
```

```python
#plt.xscale("log")
#plt.yscale("log")
plt.scatter(np.log(df["N"]), np.log(df["square_disp"]), c=df["mode"], cmap="Set1", zorder=2)
plt.xlabel(r"$\log(N)$")
plt.ylabel(r"$\log(\left<x^2\right>)$")
x = np.log(df["N"])
y = np.log(df["square_disp"])
b, a = np.polyfit(x, y, 1)
plt.plot(x, a + b*x, zorder=1)
plt.legend(["Fit mit Steigung %f" % (b), "grau=abzählen, rot=einfaches Monte-Carlo"])
plt.title("self-avoiding RW")
plt.savefig("saw.pdf")
```
